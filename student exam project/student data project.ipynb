```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student data.csv")
print (df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop unnamed column
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
#changeing weekly study hours column
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5 - 10")
df.head()
```

```python
#gender distribution
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we have analysed that:
#the number of females in the data is more than the number of males

```python
gb =  df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot = True)
plt.title("Education Difference Between Parents")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have good impact on their scores
```

```python
gb1 =  df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot = True)
plt.title("Parent Marital Status")
plt.show()
```

```python
#from the above chart we have concluded that there is neligible impact on the student score
```

#Distribution of Ethnic Groups

```python
groupA = df.loc[df['EthnicGroup'] == "group A"].shape[0]
groupB = df.loc[df['EthnicGroup'] == "group B"].shape[0]
groupC = df.loc[df['EthnicGroup'] == "group C"].shape[0]
groupD = df.loc[df['EthnicGroup'] == "group D"].shape[0]
groupE = df.loc[df['EthnicGroup'] == "group E"].shape[0]

l = ['group A', 'group B', 'group C', 'group D', 'group E']

mlist = [groupA, groupB, groupC, groupD, groupE]
print(mlist)
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
